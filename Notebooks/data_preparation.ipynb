```python
# import libraries 
import _sqlite3
import pandas as pd
```

```python
# Path to the SQLite database file
file_path = '/Users/rango/Documents/projects/SoccerForecast/database.sqlite'
try:
    # Connect to the SQLite database
    conn = _sqlite3.connect(file_path)
    print("Connection successful!")
except _sqlite3.Error as e:
    print("Error connecting to SQLite database:", e)
```

```python
match_df = pd.read_sql_query("SELECT * from Match", conn)
```

```python
#checking for dups
is_dup = (~match_df.duplicated()).all()
print(is_dup)
```

```python
#percentage of missing values for each column 
percent_missing = match_df.isnull().sum() * 100 / len(match_df)
missing_value_df = pd.DataFrame({'column_name': match_df.columns,'percent_missing': percent_missing})
missing_value_df.sort_values('percent_missing', inplace=True, ascending=False)
print(missing_value_df)
conn.close()
```
